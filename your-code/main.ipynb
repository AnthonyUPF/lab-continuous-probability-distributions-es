### Usando Numpy

```python
# Parámetros de la distribución uniforme
a = 0  # límite inferior
b = 1  # límite superior
size = 1000  # número de muestras

# Generar números aleatorios uniformes usando numpy
uniform_data_np = np.random.uniform(a, b, size)

# Visualizar la distribución usando seaborn
sns.histplot(uniform_data_np, bins=30, kde=True)
plt.title('Distribución Uniforme usando Numpy')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```

### Usando Scipy

```python
from scipy.stats import uniform

# Generar números aleatorios uniformes usando scipy
uniform_data_scipy = uniform.rvs(loc=a, scale=b-a, size=size)

# Visualizar la distribución usando seaborn
sns.histplot(uniform_data_scipy, bins=30, kde=True)
plt.title('Distribución Uniforme usando Scipy')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# your code here

def generate_exponential_data(mean, size):
    scale = mean
    exponential_data = np.random.exponential(scale, size)
    return exponential_data
```

Supongamos que el tiempo que uno pasa en un banco se distribuye exponencialmente con media de 10 minutos (es decir, λ = 1/10). ¿Cuál es la probabilidad de que un cliente pase menos de quince minutos en el banco?

### Distribución Exponencial y CDF

La distribución exponencial modela el tiempo entre eventos en un proceso de Poisson, donde los eventos ocurren de manera continua e independiente en el tiempo. Para una variable aleatoria $ X $ que sigue una distribución exponencial con parámetro $ \lambda $:

- La función de densidad de probabilidad (PDF) está dada por:
  $$
  f(x) = \lambda e^{-\lambda x} \quad \text{para } x \geq 0
  $$
  donde $ \lambda $ es la tasa de ocurrencia de eventos por unidad de tiempo.

- La función de distribución acumulativa (CDF), que da la probabilidad de que $ X $ sea menor o igual que un cierto valor $ x $, se define como:
  $$
  F(x) = P(X \leq x) = 1 - e^{-\lambda x} \quad \text{para } x \geq 0
  $$

### Explicación de la Fórmula

- **Exponencial Negativo**: La expresión $ e^{-\lambda x} $ es la probabilidad de que $ X $ sea mayor que $ x $ (es decir, $ P(X > x) $).

- **Complemento a 1**: Por lo tanto, $ 1 - e^{-\lambda x} $ es la probabilidad de que $ X $ sea menor o igual que $ x $ (es decir, $ P(X \leq x) $ o $ F(x) $).


¿Cuál es la probabilidad de que el cliente pase más de 15 minutos?

```python
# your code here

# Tiempo límite en minutos (menos de 15 minutos)
time_limit = 15

# Calcular la probabilidad usando la CDF de la distribución exponencial
probability_less_than_15 = 1 - np.exp(-lambda_param * time_limit)

print(f"La probabilidad de que un cliente pase menos de quince minutos en el banco es: {probability_less_than_15:.4f}")
```
